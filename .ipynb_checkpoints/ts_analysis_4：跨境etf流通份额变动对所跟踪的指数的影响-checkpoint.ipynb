- 有较多的指数，每个指数有不唯一的etf跟踪（https://www.jisilu.cn/data/qdii/#qdiia）
    - 应该考虑某指数的全部跟踪etf份额（每个etf在t时刻的份额变动方向和大小不一致，需要综合考虑）
    - 应该考虑多个指数（指数之间存在差异（较小））
    - 选择了恒生科技指数、恒生指数、恒生国企指数分析（跟踪上述指数的etf较多）
- 计算了“基金份额变动占比因子”，t时刻上（四点半以后），指数i的该因子计算方法是：
    - 将跟踪指数的所有etf的流通份额变动结合收盘价，计算整体金额
    - 将该金额按照指数成分权重分配给指数成分股，分别计算每个成分股分配到的金额占各自当日成交额的比重
    - 得到所有成分股的比重后，再按照成分股权重加权平均，即可计算得到“跟踪指数的所有etf的流通份额变动占指数成分股成交额的比重”
- 时序分析
    - 获取指数价格数据后，按照半小时的颗粒度计算收益（使用open价），另外还计算了上午、下午、全天的收益，便于分析“基金份额变动占比因子”对指数收益的影响是否因为时间段而不同
    - “基金份额变动占比因子”对“指数期间收益”的时序回归除了正常的OLS作为第一个模型之外，也添加时间趋势项trend作为控制变量作为第二个模型，以识别时间趋势的影响
    - 发现不同的时间段上，“基金份额变动占比因子”对收益的影响方向大多是正的，但回测区间内收益时序存在明显的负向时间趋势；后者的存在干扰了在时序上投资该因子的纯粹性（在截面上逻辑更合理）
    - 不同指数的时序分析在不同时间段上，参数大小符号和显著性存在差异
- 面板分析
    - 为了应对时序分析中矛盾的表现，考虑了面板回归。此外，察觉到明显的时间趋势，面板回归中应该更关注时间固定效应
    - 面板为T>>n的长面板，应该使用时间趋势项控制时间固定效应，而非one-hot形式（避免过多的自由度损失）
    - 观测到和时序分析基本一致的结论：“基金份额变动占比因子”对收益的影响方向大多是正的，但回测区间内收益时序存在明显的负向时间趋势
- 截面因子收益模拟
    - 为了保证“基金份额变动占比因子”投资纯粹性，在截面上做多该因子较大者，做空该因子较小者，进行收益模拟
    - 半小时颗粒度上，该因子表现并不完全一致，15:30-16:10期间段因子表现亮眼；上午、下午、全天的收益模拟中，该因子收益都优于同期买入持有指数的收益，

```python
def clean_data(etf_code, index_code):
    # etf份额变动信息，添加份额对应的金额信息'amont'
    etf_total_amont = pd.DataFrame()
    for code in etf_code:
        # etf份额变动信息，添加份额对应的金额信息'amont'
        unit_df = pd.read_excel('./qdii/{}.xlsx'.format(code), header=3)
        bar_df = pd.read_excel('./qdii/{}bar.xlsx'.format(code)).iloc[:-2]
        bar_df['Date'] = bar_df['日期'].map(lambda x: datetime.datetime(x.year, x.month, x.day))
        bar_df.rename(columns={'代码':'code', '名称':'name', '日期':'Timestamp', 
                   '开盘价(元)':'etf_open', '最高价(元)':'etf_high', '最低价(元)':'etf_low', '收盘价(元)':'etf_close'}, inplace = True) 
        d_close_pri = bar_df.groupby('Date').tail(1)[['etf_close', 'Date']]
        unit_close_df = pd.merge(unit_df, d_close_pri, on=['Date'])
        unit_close_df['{}_amont'.format(code)] = unit_close_df['etf_close'] * unit_close_df['unit_floortradingchange']
        unit_close_df = unit_close_df.iloc[:-1]
        unit_close_df = unit_close_df.set_index('Date', drop=True).rename_axis(None)[['{}_amont'.format(code)]]
        etf_total_amont = pd.merge(etf_total_amont, unit_close_df, left_index=True,right_index=True, how='outer')

    etf_total_amont['etfs_amont'] = etf_total_amont.fillna(0).sum(axis=1)

    # 获取指数成分股名称和权重信息
    start_date = etf_total_amont.index[0]  # 开始日期
    end_date = etf_total_amont.index[-1]  # 结束日期
    fields = "indexconstituent"  # 获取成分股信息的指标代码

    weight = pd.DataFrame()
    for date in tqdm_notebook(etf_total_amont.index):
        index=w.wset(fields,"date={};windcode={}".format(date, index_code),usedf=True)
        index_t = index[1][['i_weight']]
        index_t.index = index[1]['wind_code'].values
        index_t.columns = [date]
        weight = pd.merge(weight, index_t, how='outer',left_index=True,right_index=True)

    weight = weight.T
    weight.fillna(0.0, inplace=True)
    return etf_total_amont, weight

def cal_etf_chg_amt_pct(etf_total_amont, weight):
    for date in tqdm_notebook(etf_total_amont.index):
        # t时刻的etf amt和成分股权重
        amont_t = etf_total_amont.loc[date]['etfs_amont']
        weight_t = weight.loc[date]
        # 按权重分配etf amt
        weight_amont_t = pd.DataFrame(weight_t * amont_t / 100)
        weight_amont_t.columns = ['etf_chg_amt']
        weight_t = pd.DataFrame(weight_t/100)
        weight_t.columns = ['weight']
        
        # 截面读成分股“当日amt”，汇总成分股amt、etf amt、成分股权重信息
        stk_ohlc_t=w.wss(weight_amont_t.index.tolist(), "amt","tradeDate={}".format(date),usedf=True)[1]
        index_amt = sum(weight_t['weight']*stk_ohlc_t['AMT'].fillna(0))  ### 
        etf_total_amont.loc[date, 'index_amt'] = index_amt

        stk_etf_chg_amt = pd.merge(stk_ohlc_t, weight_amont_t,left_index=True,right_index=True)
        stk_etf_chg_amt = pd.merge(stk_etf_chg_amt, weight_t,left_index=True,right_index=True)
        
        stk_etf_chg_amt = stk_etf_chg_amt[(stk_etf_chg_amt['AMT']!=0) & (stk_etf_chg_amt['AMT'].notna())]
        stk_etf_chg_amt['etf_chg_amt_pct'] = abs(stk_etf_chg_amt['etf_chg_amt']) / stk_etf_chg_amt['AMT']  # 避免分母是0是空值的情况
        etf_chg_amt_pct = sum(stk_etf_chg_amt['weight'] * stk_etf_chg_amt['etf_chg_amt_pct'])
        etf_total_amont.loc[date, 'etf_chg_amt_pct'] = etf_chg_amt_pct
        
    etf_total_amont = etf_total_amont[etf_total_amont['index_amt']!=0]
    etf_total_amont['etf_chg_amt_pct'] = np.sign(etf_total_amont['etfs_amont']) * etf_total_amont['etf_chg_amt_pct']
    etf_total_amont['etf_chg_amt_pct_lag1'] = etf_total_amont['etf_chg_amt_pct'].shift()
    return etf_total_amont

def plot_pctchg_head_price(etf_total_amont, index_code):
    # “etf 份额变动 对应 指数成分股成交量 的比重”
    print('前10百分比份额变动：', etf_total_amont.sort_values(by=['etf_chg_amt_pct'],ascending=False).head(10)['etf_chg_amt_pct'].tolist())
    # 对应的下一天
    print('对应的下一日分别是：', etf_total_amont.sort_values(by=['etf_chg_amt_pct_lag1'],ascending=False).head(10).index.tolist())
    bar_df = pd.read_excel('./qdii/{}.xlsx'.format(index_code)).iloc[:-2]
    bar_df['Date'] = bar_df['日期'].map(lambda x: datetime.datetime(x.year, x.month, x.day))
    bar_df.rename(columns={'代码':'code', '名称':'name', '日期':'Timestamp', 
               '开盘价(元)':'open', '最高价(元)':'high', '最低价(元)':'low', '收盘价(元)':'close'}, inplace = True) 

    unit_pct_change_head10 = etf_total_amont.sort_values(by=['etf_chg_amt_pct_lag1'],ascending=False).index[:9]
    fig, ax = plt.subplots(3, 3, figsize=(18, 9))

    i=0
    for date in unit_pct_change_head10:
        bar_df_temp = bar_df[bar_df['Date'] == date].set_index('Timestamp')
        i += 1
        plt.subplot(int(str(33)+str(i)))
        plt.plot(bar_df_temp.index, bar_df_temp['close'])
```

```python
from statsmodels.tsa.stattools import adfuller # ADF

def augmented_dickey_fuller_test(seq, p_thres):
    [t,p,c,r]=adfuller(seq, regression='ct', regresults=True)
    
    if (p <= p_thres) & (r.resols.pvalues[-1] <= p_thres):
        print('The time series is trend-stationary.')
        is_stationary = False
    else:
        [t,p,c,r]=adfuller(seq, regression='c', regresults=True)
        if (p <= p_thres) & (r.resols.pvalues[-1] <= p_thres):
#             print('The time series is drift-stationary.')
            is_stationary = True
        else:
            [t,p,c,r]=adfuller(seq, regression='nc', regresults=True)
            if p <= p_thres:
#                 print('The time series is ordinary-stationary.')
                is_stationary = True
            else:
#                 print('The time series is not stationary.')
                is_stationary = False
                
    return is_stationary

def integrated_order_test(seq, p_thres):
    is_stationary = augmented_dickey_fuller_test(seq, p_thres)
    integrated_order = 0
    while not is_stationary:
        seq = seq.diff().dropna()
        is_stationary = augmented_dickey_fuller_test(seq, p_thres)
        integrated_order += 1
    print('The time series is integrated of order {}.'.format(integrated_order))
    return integrated_order

def cointegration_test(y_seq, x_seq, p_thres):
    y_integrated_order = integrated_order_test(y_seq, p_thres)
    x_integrated_order = integrated_order_test(x_seq, p_thres)
    
    if (min(y_integrated_order, x_integrated_order) != 0) & (y_integrated_order == x_integrated_order):
        model = sm.OLS(endog=y_seq, exog=x_seq)
        result = model.fit()
        resid = result.resid
        is_stationary = augmented_dickey_fuller_test(resid, p_thres)
        if is_stationary:
            print('Pass the cointegration test.')
            is_cointegrated = True
    else:
        is_cointegrated = False
    
    return is_cointegrated
```

```python
def ts_regression_test_periods(index_code, start_end_lst, etf_total_amont):
    bar_df = pd.read_excel('./qdii/{}.xlsx'.format(index_code)).iloc[:-2]
    bar_df['Date'] = bar_df['日期'].map(lambda x: datetime.datetime(x.year, x.month, x.day))
    bar_df.rename(columns={'代码':'code', '名称':'name', '日期':'Timestamp', 
               '开盘价(元)':'open', '最高价(元)':'high', '最低价(元)':'low', '收盘价(元)':'close'}, inplace = True) 
    
    def resample(df_grouped, start, end):
        df_grouped = df_grouped[(df_grouped['Timestamp'].map(lambda x:x.hour*100+x.minute)>=start) & 
                                (df_grouped['Timestamp'].map(lambda x:x.hour*100+x.minute)<=end)]
        if len(df_grouped):
            symbol_date_period_price = df_grouped['open']  # 使用open价格更贴合时间设定
            symbol_date_period_ret = symbol_date_period_price.iloc[-1]/symbol_date_period_price.iloc[0] - 1
            return pd.DataFrame([symbol_date_period_ret]).T
        
    for start, end in tqdm_notebook(start_end_lst):
        print('\nperiod:', int(np.floor(start/100)),':',str(start%100).rjust(2, '0'), '-', 
              int(np.floor(end/100)),':',str(end%100).rjust(2, '0'))
        bar_resample = bar_df.groupby('Date').apply(resample, start, end)
        bar_resample.columns = ['period_ret']
        bar_resample.reset_index(inplace=True)
        bar_resample.drop(['level_1'], axis=1, inplace=True)
        bar_resample.set_index('Date', inplace=True)

        bar_unit = pd.merge(etf_total_amont, bar_resample,left_index=True,right_index=True,how='left')

        corr = np.corrcoef(bar_unit.dropna()['period_ret'], bar_unit.dropna()['etf_chg_amt_pct_lag1'])[0, 1]
        print("相关系数：", round(corr,4))
        # 平稳性和协整检验
        y_seq = bar_unit.dropna()['period_ret']
        x_seq = bar_unit.dropna()['etf_chg_amt_pct_lag1']
        
        y_is_stationary = augmented_dickey_fuller_test(y_seq, 0.1)
        x_is_stationary = augmented_dickey_fuller_test(x_seq, 0.1)
        if y_is_stationary & x_is_stationary:
            print('两时序均平稳，可直接建模')
        else:
            is_cointegrated = cointegration_test(y_seq, x_seq, 0.1)
            if is_cointegrated:
                print('两时序均不平稳，但协整，可直接建模')
            else:
                print('两时序存在不平稳，也不满足协整条件，时序回归结果可信度差')

        # 回归检验
        model = sm.OLS(y_seq, x_seq)
        result = model.fit()
        print(result.summary().tables[1])
        
        # 消去确定性趋势，再做回归检验
        bar_unit['trend'] = list(range(len(bar_unit)))
        x_seq_2 = bar_unit.dropna()['etf_chg_amt_pct_lag1', 'trend']
        
        model2 = sm.OLS(y_seq, x_seq_2)
        result2 = model2.fit()
        print(result2.summary().tables[1])
```

```python
def resample_2(df_grouped, start, end):
    df_grouped = df_grouped[(df_grouped['Timestamp'].map(lambda x:x.hour*100+x.minute)>=start) & 
                            (df_grouped['Timestamp'].map(lambda x:x.hour*100+x.minute)<=end)]
    if len(df_grouped):
        symbol_date_period_price = df_grouped['open']
        symbol_date_period_ret = symbol_date_period_price.iloc[-1]/symbol_date_period_price.iloc[0] - 1
        return pd.DataFrame([symbol_date_period_ret]).T
    
indexs_code = ["HSTECH.HI", 'HSIII.HI', 'HSCEI.HI', 'HSI.HI']

for index_code in indexs_code:
    bar_df = pd.read_excel('./qdii/{}.xlsx'.format(index_code)).iloc[:-2]
    bar_df['Date'] = bar_df['日期'].map(lambda x: datetime.datetime(x.year, x.month, x.day))
    bar_df.rename(columns={'代码':'code', '名称':'name', '日期':'Timestamp', 
               '开盘价(元)':'open', '最高价(元)':'high', '最低价(元)':'low', '收盘价(元)':'close'}, inplace = True) 

    period_ret_df = pd.DataFrame()
    for start, end in tqdm_notebook(start_end_lst):
        bar_resample = bar_df.groupby('Date').apply(resample_2, start, end)
        bar_resample.columns = ['ret_{}_{}'.format(start, end)]
        bar_resample.reset_index(inplace=True)
        bar_resample.drop(['level_1'], axis=1, inplace=True)
        bar_resample.set_index('Date', inplace=True)
        period_ret_df = pd.merge(period_ret_df, bar_resample, left_index=True,right_index=True, how='outer')
    
    period_ret_df.to_csv('{}_period_ret.csv'.format(index_code))
```

# 4 Panel Regression & PnL Simulation

```python
def cal_buy_sell_side(df_grouped):
    df_temp = pd.Series([0]*len(df_grouped))
    df_temp.index = df_grouped.index
    
    min_index_code = df_grouped[df_grouped == df_grouped.min()].index[0]
    max_index_code = df_grouped[df_grouped == df_grouped.max()].index[0]
    
    df_temp.loc[min_index_code] = -0.5
    df_temp.loc[max_index_code] = 0.5
    
    return df_temp

def evalute_cumret(cumret_seq):
    day_ret = pow(cumret_seq.iloc[-1],1/len(cumret_seq)) - 1
    year_ret = pow(1+day_ret,240) - 1
    peak = np.maximum.accumulate(cumret_seq)
    drawdowns = (peak - cumret_seq) / peak
    max_drawdown = np.amax(drawdowns)
    
    year_std = cumret_seq.std() * np.sqrt(252)
    
    print('累计收益：', round((cumret_seq.iloc[-1]-1)*100,2), '%')
    print('年化收益：', round(year_ret*100,2), '%')
    print('最大回撤：', round(max_drawdown*100,2), '%')
    print('shape ratio：', round((year_ret-0.02)/year_std, 2))
    print('收益回撤比：', round((cumret_seq.iloc[-1]-1)/max_drawdown,2))
```

```python
# 获得面板数据
# indexs_code = ["HSTECH.HI", 'HSIII.HI', 'HSCEI.HI', 'HSI.HI']
indexs_code = ["HSTECH.HI", 'HSCEI.HI', 'HSI.HI']

index_period_ret_etfs_chg_panel = pd.DataFrame()
for index_code in indexs_code:
    index_period_ret = pd.read_csv('{}_period_ret.csv'.format(index_code), index_col=0)
    index_etfs_amont = pd.read_csv('{}_etfs_amont.csv'.format(index_code), index_col=0)[['etf_chg_amt_pct_lag1']]
    index_period_ret_etfs_chg = pd.merge(index_etfs_amont, index_period_ret, 
                                         left_index=True, right_index=True, how='outer').reset_index()
    index_period_ret_etfs_chg['index_code'] = index_code
    index_period_ret_etfs_chg_panel = pd.concat([index_period_ret_etfs_chg_panel, index_period_ret_etfs_chg],ignore_index=True)
```

```python
index_period_ret_etfs_chg_panel['index'] = pd.to_datetime(index_period_ret_etfs_chg_panel['index'])
for start, end in tqdm_notebook(start_end_lst):
    print('\nperiod:', int(np.floor(start/100)),':',str(start%100).rjust(2, '0'), '-', 
          int(np.floor(end/100)),':',str(end%100).rjust(2, '0'))
    panel_data = index_period_ret_etfs_chg_panel[[
        'index_code','index','ret_{}_{}'.format(start, end),'etf_chg_amt_pct_lag1']].dropna()
    
    trend_df = pd.DataFrame()
    trend_df.index = sorted(list(set(panel_data['index'])))
    trend_df['trend'] = list(range(len(set(panel_data['index']))))
    panel_data = pd.merge(panel_data, trend_df, left_on = 'index', right_index=True)  # 适配smf.ols接口
    
    panel_data_2 = panel_data.set_index(['index_code', 'index'])
    dependent_variable = panel_data_2['ret_{}_{}'.format(start, end)]
    exogenous_variables = panel_data_2[['etf_chg_amt_pct_lag1','trend']]  # 适配linearmodels.panel.PanelOLS接口
    
    # 采用ols估计，加入个体和时间的虚拟变量，本回归可以查看固定效应的虚拟变量的回归系数
#     res_ols = smf.ols('ret_{}_{}~etf_chg_amt_pct_lag1+index_code+C(index)'.format(start, end), data=panel_data)
#     res_ols = smf.ols('ret_{}_{}~etf_chg_amt_pct_lag1+index_code'.format(start, end), data=panel_data)
#     res_ols = smf.ols('ret_{}_{}~etf_chg_amt_pct_lag1+C(index)'.format(start, end), data=panel_data)

#     res_ols = smf.ols('ret_{}_{}~etf_chg_amt_pct_lag1+trend'.format(start, end), data=panel_data)
#     results_ols = res_ols.fit()
#     print(results_ols.summary().tables[1])
    
    # 混合回归：长面板控制时间效应的方法是使用时间趋势项，而不是虚拟变量，此时使用混合回归即可
    model = PanelOLS(dependent_variable, exogenous_variables, entity_effects=False, time_effects=False)  
    model_result = model.fit(cov_type='clustered', cluster_entity=True)  
    print(model_result.summary.tables[1])

#     # 个体固定效应
#     model = PanelOLS(dependent_variable, exogenous_variables, entity_effects=True, time_effects=False)  
#     model_result = model.fit(cov_type='clustered', cluster_entity=True)  
#     print(model_result.summary.tables[1])

#     # 时间固定效应
#     model = PanelOLS(dependent_variable, exogenous_variables, entity_effects=False, time_effects=True)  
#     model_result = model.fit(cov_type='clustered', cluster_entity=True)  
#     print(model_result.summary.tables[1])
    
#     # 双向固定效应
#     model = PanelOLS(dependent_variable, exogenous_variables, entity_effects=True, time_effects=True)  
#     model_result = model.fit(cov_type='clustered', cluster_entity=True) 
#     print(model_result.summary.tables[1])
    
    factor = index_period_ret_etfs_chg_panel[['index', 'index_code', 'etf_chg_amt_pct_lag1']]
    factor = factor.pivot(index = 'index', columns = 'index_code', values= 'etf_chg_amt_pct_lag1')
    period_ret = index_period_ret_etfs_chg_panel[['index', 'index_code', 'ret_{}_{}'.format(start, end)]]
    period_ret = period_ret.pivot(index = 'index', columns = 'index_code', values= 'ret_{}_{}'.format(start, end))
    
    factor = factor.loc[sorted(list(set(factor.dropna().index) & set(period_ret.dropna().index)))]
    factor_rank = factor.rank(axis=1)
    factor_buy_sell = factor_rank.apply(cal_buy_sell_side, axis=1)
    
    period_ret = period_ret.loc[sorted(list(set(factor.dropna().index) & set(period_ret.dropna().index)))]
    strategy_ret = factor_buy_sell * period_ret
    strategy_cum_ret = np.cumprod(strategy_ret.sum(axis=1) + 1)
    
    HSTECH_cumret = np.cumprod((period_ret + 1))['HSTECH.HI']
    HSCEI_cumret = np.cumprod((period_ret + 1))['HSCEI.HI']
    HSI_cumret = np.cumprod((period_ret + 1))['HSI.HI']
    
    # 评价指标计算
    print('\n策略评价指标')
    evalute_cumret(strategy_cum_ret)
    print('\nHSTECH评价指标')
    evalute_cumret(HSTECH_cumret)
    print('\nHSCEI评价指标')
    evalute_cumret(HSCEI_cumret)
    print('\nHSI评价指标')
    evalute_cumret(HSI_cumret)
    
    # plot
    plt.plot(strategy_cum_ret.index, strategy_cum_ret, label = 'strategy_{}_{}_cumret'.format(start, end))
    plt.plot(np.cumprod((period_ret + 1)).index, np.cumprod((period_ret + 1))['HSTECH.HI'], label = 'HSTECH.HI')
    plt.plot(np.cumprod((period_ret + 1)).index, np.cumprod((period_ret + 1))['HSCEI.HI'], label = 'HSCEI.HI')
    plt.plot(np.cumprod((period_ret + 1)).index, np.cumprod((period_ret + 1))['HSI.HI'], label = 'HSI.HI')
    plt.show()
```
